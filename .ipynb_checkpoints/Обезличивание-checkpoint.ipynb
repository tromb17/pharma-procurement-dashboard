```python
import os
import pandas as pd #для форматирования данных
import numpy as np #для математических расчетов
from pathlib import Path
import re
from datetime import date, timedelta, datetime
import locale
from decimal import Decimal, ROUND_UP
import warnings
from io import BytesIO
from pandas.tseries.offsets import MonthEnd
from dateutil.relativedelta import relativedelta

warnings.filterwarnings('ignore')

locale.setlocale(locale.LC_TIME, 'ru')
pd.set_option('display.max_columns', None)

pd.options.display.float_format ='{:.2f}'.format
```

```python
df_1 = pd.read_excel('Потребность 2026.xlsx', sheet_name='Sheet1')
df_1
```

```python
columns_config = {
    "Производитель": "Производитель",
    "Номенклатура.Код": "Код",
    "Наименование полное": "Препарат",
    #"МНН": "МНН"
}

# Здесь будут лежать словари соответствий, если вдруг надо
mappings = {}

for col, prefix in columns_config.items():
    if col in df_1.columns:
        # Берём уникальные значения в этом столбце
        unique_vals = (
            df_1[col]
            .astype("string")           # приведение к строке, чтобы всё было единообразно
            .drop_duplicates()
            .reset_index(drop=True)
        )

        # Строим mapping: оригинальное значение → Префикс_номер
        mapping = {val: f"{prefix}_{i+1}" for i, val in enumerate(unique_vals)}

        # Применяем mapping к столбцу
        df_1[col] = df_1[col].astype("string").map(mapping)

        # Сохраняем mapping (на всякий случай; можно не хранить, если не нужно)
        mappings[col] = mapping
```

```python
def scale_amount_columns_inplace(df, columns, low=0.6, high=0.9, seed=42):
    """
    Масштабирует УЖЕ существующие числовые столбцы inplace.
    Один общий случайный коэффициент для всех указанных столбцов.
    """
    rng = np.random.default_rng(seed=seed)
    k = rng.uniform(low, high)

    for col in columns:
        if col in df.columns:
            df[col] = df[col].astype(float) * k

    return df, k  # k можно сохранить отдельно, если нужно

# пример использования
cols_to_scale = ["Потребность, руб."]#, "В заявке, руб.", "Потребность, руб."]#, "Потребность, руб.", "Цена за единицу"]
df_1, k_used = scale_amount_columns_inplace(df_1, cols_to_scale, low=0.6, high=0.9, seed=42)
```
